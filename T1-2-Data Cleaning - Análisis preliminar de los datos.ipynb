# Variables dummy

Las variables de ayuda, o variables artificilaes, etc. Es un proceso que crea variables separadas para cada categoria en el caso de tener una variable categorica, apesar que las varoables categoricas contienen gran cantidad de informacion, rn ocaciones es util mostrar una relacion de causalidad con una sola variable de salida en un modelo predictivo lineal, regrecion logistica o de clastering sin necesidad de procesarlo, es decir comvertir esa categoria en una determinada columna.

Por ejemplo la variable sex es una variable categorica definbe una categoria una clasificacion de cada una de las filas por ¨female¨ o ¨male¨, se crearan dos variables dummy, una para la columna male que tendra un 1 en casa de ser mele y en caso contrario un 0, de igual forma susedera con la otra columna female.

Se definen tantas columnas como categorias se tengan.

## Funcion para  generar variables Dummys 

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy ], axis = 1)
    return df
```
